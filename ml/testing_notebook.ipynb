```python
import pandas as pd
from pathlib import Path

# Get all CSV files from binary_search_results directory
results_dir = Path("../models/binary_search_results")
csv_files = sorted(results_dir.glob("binary_search_*.csv"))

# Read and concatenate all CSV files
dfs = [pd.read_csv(f) for f in csv_files]
df = pd.concat(dfs, ignore_index=True)

# Create df_1 as filtered version
df_1 = df.drop(columns=['met_precision_req',
       'roc_auc', 'pr_auc', 'true_positives', 'false_positives', 'total_samples',
       'buy_samples', 'buy_pct', 'win_rate'])

df_1 = df_1[['f1', 'precision', 'recall', 'window_size', 'horizon', 'buy_threshold', 'classifier',
         'n_estimators', 'max_depth', 'min_precision', 'threshold', 'accuracy']]

df_1 = df_1[(df_1['f1'] > 0.5) & (df_1['recall'] < 0.4) & (df_1['recall'] > 0.1)]

df_1 = df_1.sort_values(by='precision', ascending=False)

print(f"Loaded {len(csv_files)} files with {len(df)} total rows and {len(df_1)} filtered rows")

df_1.head()
```
